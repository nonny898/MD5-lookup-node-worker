```python
import pandas as pd
import hashlib
```

```python
dataframe = pd.read_csv('rockyou.txt',
                        delimiter = "\n", 
                        header = None, 
                        names = ["Passwords"],
                        encoding = "ISO-8859-1")
```

```python
dataframe['MD5'] = [hashlib.md5(str.encode(str(i))).hexdigest() 
                    for i in dataframe['Passwords'].fillna(0).astype(str)]
```

```python
dataframe
```

```python
columnsTitles=["MD5","Passwords"]

dataframe=dataframe.reindex(columns=columnsTitles)

dataframe
```

```python
dataframe.to_csv(r'hashed.txt', header=None, index=None, sep=' ')
```
